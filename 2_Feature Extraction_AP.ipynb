```python
########################################################################################
## 1. Active protein
data = pd.read_csv("Input/nram_actives50_ap.csv")
data
```

```python
########################################################################################
## 1.a pubchem
fps = [get_fingerprint(x, fp_type='pubchem') for x in data.smiles]
fps
```

```python
#menyiapkan dataframe kosong sesuai jumlah bit binary tipe fingerprint
df_hasil = pd.DataFrame([0]*881).T #ubah sesuai jumlah bit tipe fineprint

#memasukkan data bit binary ke dalam dataframe
for x in fps:
    hasil = [0]*881
    #mengubah 0 menjadi 1 sesuai lokasi fingerprint
    for i in x:
        hasil[(i-1)] = 1
        df1 = pd.DataFrame(hasil).T
    df_hasil = df_hasil.append(df1, ignore_index = True)
df_hasil = df_hasil.drop(0)
    
df_hasil_real = df_hasil.copy()
df_hasil = pd.DataFrame([0]*881).T

#agar index sama dengan data asli
df_hasil_real = df_hasil_real.reset_index()
df_hasil_real = df_hasil_real.drop("index", axis=1)
df_hasil_real.head()
```

```python
## ERROR
##join with data aslinya
#data_akhir = data.merge(df_hasil, left_index=True, right_index=True)
#data_akhir.head()
```

```python
#join with data aslinya
data_akhir = data.merge(df_hasil_real, left_index=True, right_index=True)
data_akhir.head()
```

```python
#save ke csv
data_akhir.to_csv('Output/fitur_nram_actives50_pubchem.csv', sep=',', index=False)
```

```python
########################################################################################
## 1. Active protein
data = pd.read_csv("Input/nram_actives50_ap.csv")
data
```

```python
########################################################################################
## 1.b klekota-roth
fps = [get_fingerprint(x, fp_type='klekota-roth') for x in data.smiles]
fps
```

```python
#menyiapkan dataframe kosong sesuai jumlah bit binary tipe fingerprint
df_hasil = pd.DataFrame([0]*4860).T #ubah sesuai jumlah bit tipe fineprint

#memasukkan data bit binary ke dalam dataframe
for x in fps:
    hasil = [0]*4860
    #mengubah 0 menjadi 1 sesuai lokasi fingerprint
    for i in x:
        hasil[(i-1)] = 1
        df1 = pd.DataFrame(hasil).T
    df_hasil = df_hasil.append(df1, ignore_index = True)
df_hasil = df_hasil.drop(0)
    
df_hasil_real = df_hasil.copy()
df_hasil = pd.DataFrame([0]*4860).T

#agar index sama dengan data asli
df_hasil_real = df_hasil_real.reset_index()
df_hasil_real = df_hasil_real.drop("index", axis=1)
df_hasil_real.head()
```

```python
#join with data aslinya
data_akhir = data.merge(df_hasil_real, left_index=True, right_index=True)
data_akhir.head()
```

```python
#save ke csv
data_akhir.to_csv('Output/fitur_nram_actives50_klekota-roth.csv', sep=',', index=False)
```
